# Split the train and validation images

```python
import os
import random
import shutil
```

```python
source_folder = "data/Original/train/"
```

```python
folder1 = "data/train_images/"
folder2 = "data/val_images/"
os.makedirs(folder1, exist_ok=True)
os.makedirs(folder2, exist_ok=True)
```

```python
all_files = os.listdir(source_folder)

total_files = len(all_files)
percent_folder1 = 0.8  # 80%
num_files_folder1 = int(total_files * percent_folder1)
num_files_folder2 = total_files - num_files_folder1
```

```python
len(all_files)
```

```python
total_files = len(all_files)
percent_folder1 = 0.8  # 80%
num_files_folder1 = int(total_files * percent_folder1)
num_files_folder2 = total_files - num_files_folder1
```

```python
random.shuffle(all_files)
```

```python
# Copy files to folder-1
for file_name in all_files[:num_files_folder1]:
    source_path = os.path.join(source_folder, file_name)
    destination_path = os.path.join(folder1, file_name)
    shutil.copy(source_path, destination_path)
    print(f"Copied {file_name} to {folder1}")
```

```python
for file_name in all_files[num_files_folder1:]:
    source_path = os.path.join(source_folder, file_name)
    destination_path = os.path.join(folder2, file_name)
    shutil.copy(source_path, destination_path)
    print(f"Copied {file_name} to {folder2}")
```

# Prepare Validation Masks dataset

```python
import os
import shutil
```

```python
source_folder = "data/val_images/"
folder_to_check = "data/Original/train_masks/"
destination_folder = "data/val_masks/"
```

```python
file_names_to_copy = os.listdir(source_folder)
```

```python
len(file_names_to_copy)
```

```python
for file_name in file_names_to_copy:
    file_name = file_name.replace(".jpg", "").strip()
    if os.path.exists(os.path.join(folder_to_check, f"{file_name}_mask.gif")):
        source_path = os.path.join(folder_to_check, f"{file_name}_mask.gif")
        destination_path = os.path.join(destination_folder, f"{file_name}_mask.gif")

        shutil.copy(source_path, destination_path)
        print(f"Copied {file_name}_mask.gif to {destination_folder}")
```
